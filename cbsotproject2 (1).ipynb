```python

!pip install --upgrade --force-reinstall transformers==4.51.3 huggingface_hub==0.36.2
from gliner import GLiNER

model = GLiNER.from_pretrained("urchade/gliner_medium-v2.1")


labels = [
    "programming language", "framework", "library", "model",
    "organization", "metric", "dataset", "hardware",
    "technique", "methodology", "research field", "mathematical concept"
]

def extract_all_entities(text):

    entities = model.predict_entities(text, labels, threshold=0.2)

    # Result formatting
    results = {}
    for entity in entities:
        label = entity["label"]
        if label not in results:
            results[label] = []
        if entity["text"] not in results[label]:
            results[label].append(entity["text"])
    return results

```

```python
paper_text = "We implemented the model in PyTorch using a transformer architecture on the C4 dataset."
print(extract_all_entities(paper_text))
```

```python
import pandas as pd

def display_entity_table(entities_dict):

    data = []
    for label, items in entities_dict.items():
        for item in items:
            data.append({"Category": label, "Entity": item})

    if not data:
        print("No entities found.")
        return


    df_entities = pd.DataFrame(data)


    print("\n--- Identified Technical Entities ---")
    print(df_entities.to_string(index=False))
```

```python

sample_text = "PyTorch is a framework for deep learning models using BERT architecture."


entities = extract_all_entities(sample_text)


display_entity_table(entities)
```

```python
def search_and_analyze_papers(query, filter_keyword=None, k=3):
    # 1. Retrieval
    query_embedding = model.encode([query])
    faiss.normalize_L2(query_embedding)
    D, I = index.search(query_embedding, k * 5)
    results_found = 0
    print(f"\n🚀 Analysis for: '{query}'\n" + "="*60)

    for score, idx in zip(D[0], I[0]):
        paper = df.iloc[idx]

        # 2. Keyword Filter
        if filter_keyword:
            text_to_check = (str(paper['title']) + " " + str(paper['abstract'])).lower()
            if filter_keyword.lower() not in text_to_check:
                continue


        confidence_score = (1 - (score / 2)) * 100

        print(f"\n📌 Title: {paper['title']}")
        print(f"📊 Confidence: {confidence_score:.2f}%")
        print(f"📝 Abstract: {paper['abstract'][:200]}...")


        full_text = f"{paper['title']} {paper['abstract']}"
        entities_dict = extract_all_entities(full_text)

        display_entity_table(entities_dict)
        print("-" * 60)

        results_found += 1
        if results_found >= k:
            break

    if results_found == 0:
        print("No matches found with that filter.")
```
